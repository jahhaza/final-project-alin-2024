```python
import numpy as np

# Matriks A yang diberikan
A = np.array([[3, 6, -3, 5],
              [-5, -1, 7, -2],
              [9, 4, -2, 3],
              [4, -5, 6, 6]], dtype=float)

# Mendapatkan jumlah baris dan kolom dari matriks A
n = A.shape[0]

# Variabel untuk menghitung determinan
det = 1
row_swaps = 0  # untuk menghitung jumlah pertukaran baris

# Langkah 1: Operasi Baris untuk Membentuk Segitiga Atas
for i in range(n):
    # Jika elemen diagonal A[i, i] adalah 0, lakukan pertukaran baris
    if A[i, i] == 0:
        for j in range(i+1, n):
            if A[j, i] != 0:
                A[[i, j]] = A[[j, i]]  # pertukaran baris
                row_swaps += 1
                break
    
    # Jika pivot tidak sama dengan 0, lanjutkan operasi
    if A[i, i] != 0:
        # Skala baris i sehingga A[i, i] menjadi 1 (tidak wajib, tapi sering dilakukan)
        A[i] = A[i] / A[i, i]

        # Eliminasi kolom di bawah pivot untuk membentuk segitiga atas
        for j in range(i+1, n):
            if A[j, i] != 0:
                factor = A[j, i]
                A[j] = A[j] - factor * A[i]
    
    print(f"Setelah Mengolah Baris {i+1}:")
    print(np.round(A, 2))  # Menampilkan matriks setelah setiap operasi baris, dibulatkan ke 2 angka di belakang koma

# Langkah 2: Menghitung Determinan
# Determinan adalah hasil perkalian elemen diagonal utama
det = np.prod(np.diag(A))

# Jika ada pertukaran baris, kita mengalikan determinan dengan (-1) sebanyak jumlah pertukaran baris
if row_swaps % 2 != 0:
    det *= -1

# Pembulatan hasil determinan hingga 2 angka di belakang koma
det = round(det, 2)

print("\nMatriks dalam Bentuk Segitiga Atas:")
print(np.round(A, 2))  # Menampilkan matriks segitiga atas yang sudah dibulatkan

print("\nDeterminan dari Matriks A:", det)
```

```python
import numpy as np

# Matriks yang diberikan
A = np.array([[4, 6, -3, -2],
              [-8, -1, 8, -3],
              [9, 9, -2, -3],
              [4, -5, 3, 7]], dtype=float)

# Fungsi untuk menghitung determinan menggunakan eliminasi Gauss
def gauss_determinant(matrix):
    n = matrix.shape[0]
    A = matrix.copy()
    
    # Menampilkan matriks awal
    print("Matriks awal:")
    print(np.round(A, 2))
    print()

    # Eliminasi Gauss
    for i in range(n):
        # Pencarian pivot untuk baris i
        max_row = np.argmax(np.abs(A[i:n, i])) + i
        if A[max_row, i] == 0:
            return 0  # Matriks singular, determinan 0
        
        # Tukar baris jika perlu
        if max_row != i:
            A[[i, max_row]] = A[[max_row, i]]
            print(f"Pivot pada kolom {i}: Tukar baris {i} dan baris {max_row}")
            print(np.round(A, 2))
            print()

        # Eliminate entries below pivot
        for j in range(i + 1, n):
            factor = A[j, i] / A[i, i]
            A[j, i:] -= factor * A[i, i:]
            print(f"Eliminasi baris {j} dengan faktor {factor:.2f}")
            print(np.round(A, 2))
            print()

    # Determinan adalah hasil perkalian elemen diagonal utama
    determinant = np.prod(np.diagonal(A))
    return np.round(determinant, 2)

# Menghitung determinan matriks A dan mencetak langkah-langkah
det = gauss_determinant(A)
print("Determinannya adalah:", det)
```

```python
import numpy as np

# Matriks A yang terbentuk dari titik P, Q, R
A = np.array([[9, -5, 3],
              [-8, 5, -7],
              [9, -4, 7]], dtype=float)

# Fungsi untuk melakukan eliminasi Gauss dan mengubah matriks menjadi segitiga atas
def gauss_elimination_step_by_step(A):
    n = A.shape[0]
    
    for i in range(n):
        # Menampilkan matriks awal
        print(f"Langkah {i+1}: Matriks A sebelum eliminasi pada baris {i+1}")
        print(np.round(A, 2))
        print()
        
        # Membuat elemen diagonal menjadi 1 (jika diperlukan)
        if A[i, i] != 0:
            A[i] = A[i] / A[i, i]
        
        # Menampilkan matriks setelah normalisasi baris
        print(f"Setelah normalisasi baris {i+1}:")
        print(np.round(A, 2))
        print()
        
        # Eliminasi di bawah elemen diagonal
        for j in range(i + 1, n):
            factor = A[j, i]
            A[j] = A[j] - factor * A[i]
        
    return A

# Matriks A setelah eliminasi Gauss
A_upper = gauss_elimination_step_by_step(A.copy())

# Determinan adalah hasil perkalian elemen diagonal utama
determinant = np.prod(np.diagonal(A_upper))

# Menampilkan hasil
print(f"Matriks setelah eliminasi Gauss (Segitiga Atas):")
print(np.round(A_upper, 2))
print(f"Determinannya adalah: {np.round(determinant, 2)}")
```

```python
import numpy as np

# Titik-titik f, g, h
f = np.array([6, -7, 3])
g = np.array([-6, 8, -3])
h = np.array([8, -6, 3])

# Langkah 1: Tentukan vektor fg dan fh
fg = g - f
fh = h - f

print("Langkah 1: Tentukan dua vektor yang membentuk bidang:")
print(f"Vektor fg = g - f = {g} - {f} = {fg}")
print(f"Vektor fh = h - f = {h} - {f} = {fh}")
print()

# Langkah 2: Tentukan vektor normal (perkalian silang)
normal = np.cross(fg, fh)

print("Langkah 2: Tentukan vektor normal (perkalian silang antara fg dan fh):")
print(f"Vektor normal = fg x fh = np.cross({fg}, {fh}) = {normal}")
print()

# Langkah 3: Tentukan persamaan bidang
A, B, C = normal
D = -np.dot(normal, f)  # Menggunakan titik f untuk menghitung D

print("Langkah 3: Tentukan persamaan bidang:")
print(f"Vektor normal = ({A}, {B}, {C})")
print(f"Substitusi titik f = {f} ke dalam persamaan bidang untuk menghitung D:")
print(f"D = -np.dot({normal}, {f}) = {-np.dot(normal, f)}")
print(f"D = {D}")
print()

# Langkah 4: Persamaan akhir
print("Langkah 4: Persamaan bidang:")
print(f"Persamaan bidang: {A}x + ({B})y + ({C})z + ({D}) = 0")
```

```python
import numpy as np

# Diketahui vektor h + k dan h - k
h_plus_k = np.array([7, -5, 8])
h_minus_k = np.array([5, -4, 7])

# Menghitung vektor h dan k
h = (h_plus_k + h_minus_k) / 2
k = h_plus_k - h

print(f"Vektor h: {h}")
print(f"Vektor k: {k}")

# Menghitung |h + k|^2
h_plus_k_squared = np.dot(h_plus_k, h_plus_k)
print(f"||h + k||^2 = {h_plus_k_squared}")

# Menghitung |h - k|^2
h_minus_k_squared = np.dot(h_minus_k, h_minus_k)
print(f"||h - k||^2 = {h_minus_k_squared}")

# Menghitung h . k
dot_product = np.dot(h, k)
print(f"h . k = {dot_product}")
```

```python
import numpy as np

# Titik a, b, c
a = np.array([9, -5, 3])
b = np.array([-8, 5, -7])
c = np.array([9, -4, 7])

# Langkah 1: Hitung vektor ba (a - b) dan bc (c - b)
ba = a - b
bc = c - b

# Print langkah pertama
print("Langkah 1: Menghitung vektor ba dan bc")
print(f"Vektor ba (a - b) = {a} - {b} = {ba}")
print(f"Vektor bc (c - b) = {c} - {b} = {bc}")
print()

# Langkah 2: Hitung cross product ba x bc
cross_product = np.cross(ba, bc)

# Print langkah kedua
print("Langkah 2: Menghitung cross product ba x bc")
print(f"Cross product ba x bc = np.cross({ba}, {bc}) = {cross_product}")
print()

# Langkah 3: Hitung panjang vektor dari cross product
magnitude_cross_product = np.linalg.norm(cross_product)

# Print langkah ketiga
print("Langkah 3: Menghitung panjang dari cross product")
print(f"Panjang cross product = np.linalg.norm({cross_product}) = {magnitude_cross_product:.2f}")
print()

# Langkah 4: Hitung luas segitiga
area = 0.5 * magnitude_cross_product

# Print hasil akhir
print("Langkah 4: Menghitung luas segitiga")
print(f"Luas segitiga = 0.5 * {magnitude_cross_product:.2f} = {area:.2f}")
```

```python
import numpy as np

# Matriks koefisien dan hasil
A = np.array([[7, 8, -4],
              [-4, 5, -5],
              [7, -2, -6]], dtype=float)

B = np.array([-34, 14, -86], dtype=float)

# Matriks Augmented [A | B]
augmented_matrix = np.column_stack((A, B))

# Fungsi Gauss-Jordan
def gauss_jordan(matrix):
    rows, cols = matrix.shape
    for i in range(rows):
        # Langkah 1: Membuat pivot 1 pada baris ke-i
        print(f"\nLangkah {i+1}: Menjadikan elemen di baris {i+1}, kolom {i+1} sebagai pivot 1")
        matrix[i] = matrix[i] / matrix[i, i]
        matrix[i] = np.round(matrix[i], 2)  # Pembulatan hasil baris hingga 2 angka di belakang koma
        print(f"Setelah membagi baris {i+1} dengan elemen pivot:")
        print(matrix)
        
        # Langkah 2: Menghilangkan elemen di bawah dan di atas pivot
        for j in range(rows):
            if j != i:
                print(f"\nMenghilangkan elemen di baris {j+1}, kolom {i+1} dengan mengurangi {matrix[j, i]:.2f} * baris {i+1}")
                factor = matrix[j, i]
                matrix[j] = matrix[j] - factor * matrix[i]
                matrix[j] = np.round(matrix[j], 2)  # Pembulatan hasil baris hingga 2 angka di belakang koma
                print(f"Setelah operasi baris {j+1}:")
                print(matrix)
    
    return matrix

# Menerapkan Gauss-Jordan pada matriks augmented
augmented_matrix = gauss_jordan(augmented_matrix)

# Menampilkan hasil akhir
print("\nMatriks setelah Gauss-Jordan (RREF):")
print(np.round(augmented_matrix, 2))  # Pembulatan matriks akhir

# Menampilkan solusi (k1, k2, k3)
k1, k2, k3 = augmented_matrix[:, -1]
print(f"\nNilai k1: {round(k1, 2)}, k2: {round(k2, 2)}, k3: {round(k3, 2)}")
```

```python
import numpy as np

# Definisikan matriks A
A = np.array([[-1, 4, -2],
              [-3, 4, 0],
              [-3, 1, 3]])

# Langkah 1: Hitung matriks A - λI
# Untuk ini, kita akan menggunakan placeholder λ (nilai eigen) yang akan dihitung setelah persamaan karakteristik

print("Matriks A:")
print(A)

# Matriks identitas I berukuran 3x3
I = np.eye(3)

print("\nMatriks Identitas I:")
print(I)

# Langkah 2: Menyusun persamaan karakteristik det(A - λI) = 0
# Kita akan menghitung nilai eigen dengan np.linalg.eig, yang akan secara otomatis melakukan ini
eigenvalues, eigenvectors = np.linalg.eig(A)

# Langkah 3: Tampilkan nilai eigen dan vektor eigen
print("\nNilai Eigen (λ):")
print(eigenvalues)

print("\nVektor Eigen (v):")
print(eigenvectors)

# Langkah 4: Menunjukkan hubungan A * v = λ * v untuk memverifikasi hasil
for i in range(len(eigenvalues)):
    eigenvalue = eigenvalues[i]
    eigenvector = eigenvectors[:, i]
    result = np.dot(A, eigenvector)
    expected = eigenvalue * eigenvector
    print(f"\nVerifikasi untuk λ = {eigenvalue}:")
    print("A * v = ", result)
    print(f"λ * v = {expected}")
    print("Hasilnya harus sama (A * v = λ * v).")
```

```python
import numpy as np

# Mendefinisikan vektor-vektor u1, u2, u3
u1 = np.array([-4, 8, 2])
u2 = np.array([7, -3, 6])
u3 = np.array([6, -3, 7])

# Fungsi untuk menghitung proyeksi vektor u terhadap vektor v
def proj(u, v):
    return (np.dot(u, v) / np.dot(v, v)) * v

# Langkah 1: Tentukan e1
e1 = u1 / np.linalg.norm(u1)

# Langkah 2: Tentukan e2
e2 = u2 - proj(u2, e1)
e2 = e2 / np.linalg.norm(e2)

# Langkah 3: Tentukan e3
e3 = u3 - proj(u3, e1) - proj(u3, e2)
e3 = e3 / np.linalg.norm(e3)

# Menampilkan hasil
print("Basis Ortonormal:")
print("e1 =", np.round(e1, 2))
print("e2 =", np.round(e2, 2))
print("e3 =", np.round(e3, 2))
```

```python
import numpy as np

# Definisikan matriks A
A = np.array([[-6, 7, -9, 7, 2],
              [9, 4, -3, -3, 5],
              [-4, 7, -4, 2, 7],
              [7, -3, 7, 3, -5]])

# Fungsi untuk melakukan reduksi baris echelon
def gauss_elimination(A):
    rows, cols = A.shape
    r = 0  # baris pivot

    for c in range(cols):
        if r >= rows:
            break
        
        # Mencari elemen pivot (yang terbesar di kolom saat ini)
        if A[r, c] == 0:
            for i in range(r+1, rows):
                if A[i, c] != 0:
                    A[[r, i]] = A[[i, r]]  # Tukar baris
                    print(f"Tukar baris {r+1} dan {i+1}")
                    break
        
        # Jika elemen pivot tidak nol, kita lakukan eliminasi
        if A[r, c] != 0:
            # Normalisasi baris pivot
            A[r] = A[r] / A[r, c]
            print(f"Normalisasi baris {r+1}: {A[r]}")
            
            # Eliminasi elemen di bawah pivot
            for i in range(r + 1, rows):
                if A[i, c] != 0:
                    print(f"Eliminasi baris {i+1} dengan baris {r+1}")
                    A[i] = A[i] - A[i, c] * A[r]
                    print(f"Baris {i+1} setelah eliminasi: {A[i]}")
            
            r += 1
        print(f"\nMatriks setelah iterasi {r}:")
        print(A)
        print("\n------------------------------")

    return A

# Lakukan eliminasi Gauss pada matriks A
A_echelon = A.astype(float)  # Pastikan tipe data float untuk pembagian
A_echelon = gauss_elimination(A_echelon)

# Menampilkan hasil bentuk echelon baris
print("Bentuk Echelon Matriks A:")
print(A_echelon)

# Menentukan basis untuk row space (baris non-zero setelah reduksi)
basis_row_space = []
for row in A_echelon:
    if np.any(row != 0):  # Jika baris tidak nol
        basis_row_space.append(row)

# Menampilkan basis row space
print("\nBasis untuk Row Space A:")
for vector in basis_row_space:
    print(vector)
```

```python
import numpy as np

# Matriks A
A = np.array([[-6, 7, -9, 7, 2],
              [9, -4, 6, -2, 6],
              [-4, 9, -5, 7, -3],
              [7, -7, 5, 2, 8]], dtype=float)

def print_matrix_step(step, matrix):
    print(f"Step {step}:")
    print(matrix)
    print("\n")

# Menampilkan matriks awal
print_matrix_step(0, A)

rows, cols = A.shape
pivot_row = 0

# Reduksi baris echelon
for col in range(cols):
    if pivot_row >= rows:
        break
    
    # Cari baris dengan elemen terbesar di kolom saat ini
    max_row = np.argmax(np.abs(A[pivot_row:rows, col])) + pivot_row
    if A[max_row, col] != 0:
        # Tukar baris jika diperlukan
        A[[pivot_row, max_row]] = A[[max_row, pivot_row]]
        
        print_matrix_step(f"Setelah pertukaran baris di kolom {col}", A)
        
        # Normalisasi baris pivot
        A[pivot_row] = A[pivot_row] / A[pivot_row, col]
        print_matrix_step(f"Setelah normalisasi baris {pivot_row}", A)
        
        # Eliminasi elemen di bawah dan di atas pivot
        for row in range(rows):
            if row != pivot_row:
                A[row] = A[row] - A[row, col] * A[pivot_row]
        
        print_matrix_step(f"Setelah eliminasi kolom {col}", A)
        
        pivot_row += 1

# Menemukan baris yang tidak semua nol untuk basis row space
basis = []
for i in range(A.shape[0]):
    if np.any(A[i] != 0):  # Jika baris bukan baris nol
        basis.append(A[i])  # Tambahkan baris tersebut ke basis

basis = np.array(basis)

# Menampilkan basis row space
print("Basis dari Row Space A:")
print(basis)
```
